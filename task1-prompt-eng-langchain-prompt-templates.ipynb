Task 1: Replace Hardcoded Prompts

```python
from langchain_core.prompts import PromptTemplate

template_string="Explain {topic} in simple terms for beginners"
explain_prompt=PromptTemplate(
    input_variables=["topic"],
    template=template_string
)
formatted_prompt=explain_prompt.format(topic="Prompt Engineering")
print("Output:")
print(formatted_prompt)
```

Task 2: Multi-Input Prompt System

```python
multi_template_string="Explain {topic} for {audience} in a {tone} tone."

multi_prompt=PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_template_string
)

# Define test cases as a list of dictionaries
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("Outputs:")
# Loop through the test cases and generate prompts
for test in test_cases:
    result = multi_prompt.format(**test)
    print("- " + result)
```

Task 3: Prompt Variations Engine

```python
teaching_prompt=PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_prompt=PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}."
)

storytelling_prompt=PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

topic_input="Machine Learning"

print("Outputs:")
print("[Teaching]", teaching_prompt.format(topic=topic_input))
print("[Interview]", interview_prompt.format(topic=topic_input))
print("[Storytelling]", storytelling_prompt.format(topic=topic_input))
```

Task 4: ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

# Define system instructions for each role
role_instructions={
    "teacher": "You are a patient teacher. Explain concepts simply with real-world examples.",
    "interviewer": "You are a technical interviewer. Ask 3 challenging questions to test the candidate's knowledge.",
    "motivator": "You are an enthusiastic coach. Explain how mastering this topic will change the world!"
}

# Define the inputs to test
roles_to_test=["teacher", "interviewer", "motivator"]
topic_input="Neural Networks"

print("Outputs:\n")

# Loop through each role and generate the specific chat prompt
for role in roles_to_test:
    system_message=role_instructions[role]
    chat_template=ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Let's talk about {topic}.")
    ])
    formatted_chat=chat_template.format_messages(topic=topic_input)
    print("Role: " + role.upper())
    for message in formatted_chat:
        print(message.type.capitalize() + " Message: " + message.content)
    print("\n")
```

Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audiences=["beginner", "intermediate", "expert"]
    valid_tones=["formal", "casual", "fun"]
    if audience.lower() not in valid_audiences:
        print("Warning: '" + str(audience) + "' is invalid. Defaulting to 'beginner'.")
        audience="beginner"
    if tone.lower() not in valid_tones:
        print("Warning: '" + str(tone) + "' is invalid. Defaulting to 'casual'.")
        tone="casual"
    return audience, tone
test_cases=[
    {"audience": "expert", "tone": "angry"},
    {"audience": "novice", "tone": "fun"},
    {"audience": "beginner", "tone": "formal"},
    {"audience": "master", "tone": "serious"}
]

print("Outputs:\n")
for test in test_cases:
    original_audience=test["audience"]
    original_tone=test["tone"]
    print("\nTesting -> Audience: '" + original_audience + "', Tone: '" + original_tone + "'")
    safe_audience, safe_tone=validate_inputs(original_audience, original_tone)
    print("Result  -> Audience: " + str(safe_audience) + ", Tone: " + str(safe_tone))
```

Task 6: Prompt Generator App

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):
    valid_audience, valid_tone=validate_inputs(audience, tone)
    templates={
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone clearly step by step."
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask 3 questions about {topic} for an {audience} audience in a {tone} tone."
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style."
        )
    }
    selected_prompt=templates.get(style.lower(), templates["teaching"])
    return selected_prompt.format(
        topic=topic,
        audience=valid_audience,
        tone=valid_tone
    )
print("Output:")
final_prompt=generate_prompt(
    topic="Neural Networks",
    audience="beginners",
    tone="fun",
    style="storytelling"
)
print("Generated Prompt: " + final_prompt)
```

Task 7: Template Reusability Test

```python
from langchain_core.prompts import PromptTemplate

reusable_template=PromptTemplate(
    input_variables=["action", "subject"],
    template="Write a {action} about {subject}."
)
batch_inputs=[
    {"action": "poem", "subject": "the ocean"},
    {"action": "tweet", "subject": "a new Python release"},
    {"action": "short story", "subject": "a time traveler"},
    {"action": "joke", "subject": "data scientists"},
    {"action": "haiku", "subject": "coffee"}
]
print("Outputs")
for data in batch_inputs:
    output=reusable_template.format(
        action=data["action"],
        subject=data["subject"]
    )
    print("\nInput: " + data["action"] + " | " + data["subject"])
    print("Prompt: " + output)
```
